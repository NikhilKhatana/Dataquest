```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics','Physical Sciences', 'Computer Science','Engineering']
lib_arts_cats = ['Foreign Languages','English','Communications and Journalism','Art and Performance','Social Sciences and History']
other_cats = ['Health Professions','Public Administration','Education','Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(18, 20))

# First Column, STEM categories

for sp,arrin in zip(range(0,18,3), range(0,6)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[arrin]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[arrin]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(stem_cats[arrin])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    
    if sp == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2007, 10, 'Men')
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2005, 8, 'Women')
        ax.tick_params(labelbottom='on')
# Second Column, Liberal Arts Categories        
        
for sp,arrin in zip(range(1,15,3), range(0,5)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[arrin]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[arrin]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])        
    ax.set_title(lib_arts_cats[arrin])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    
    if sp == 1:
        ax.text(2005, 83, 'Women')
        ax.text(2005, 12, 'Men')
    elif sp == 13:
        ax.tick_params(labelbottom='on')
        
# Third Column, Other Categories

for sp,arrin in zip(range(2,18,3), range(0,6)):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[arrin]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[arrin]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])        
    ax.set_title(other_cats[arrin])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    
    if sp == 2:
        ax.text(2005, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif sp == 17:
        ax.text(2005, 60, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom='on')
plt.savefig('gender_degrees.png')
plt.show()
```
